### Steps to obtain a Personal Access Token (PAT):

1.  **Login to Synapse:**
    Visit [https://www.synapse.org/](https://www.synapse.org/) and log in with your account (email and password).

2.  **Access Account Settings:**
    After logging in, click on your avatar/profile in the top right corner and select **Account Settings** (or access via this direct link: [Synapse Account Settings](https://accounts.synapse.org/authenticated/myaccount?appId=synapse.org)).

3.  **Find Personal Access Tokens:**
    Scroll down to the "Personal Access Tokens" section. Click on **Manage Personal Access Tokens**.

4.  **Create a New Token:**
    * Click the **Create New Token** button.
    * **Token Name:** Enter a name (e.g., "BraTS Download Token").
    * **Token Permissions:** Select **View** (required) and **Download** (to download data). Select **Modify** only if you need to upload/edit.
    * Click **Create Token**. The token will be displayed immediately (a long string starting with `eyJ...`).

5.  **Copy Token:**
    > **Important:** Copy the token immediately as it is only shown once. Paste it into the code cell in this notebook. If you lose it, you will have to create a new one.
